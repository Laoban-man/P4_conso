# Presentation du cas

L'objectif du cas est de manipuler une base de données de telle sorte à pouvoir alimenter une application dans le domaine de la santé alimentaire. En l'occurence, l'application proposée récupère les caractéristiques d'un article scanné et propose des alternatives plus saines, en fonction du nutriscore, ayant le plus de points de similarités, notamment la catégorie et les groupes d'aliments, afin d'avoir des suggestions aussi pertinentes que possible.
La préparation des données nécessite donc d'extraire les colonnes avec des codes d'identification, les catégories d'aliments, les différentes évalutions nutritives dont le nutriscore, ainsi que les données nutritives qui pourront éventuellement être utilisée dans des analyses de régimes.

# Data exploration

```python
def quant_data_exploration(data,columns,mode="box"):
    """
    Fonction qui presente differents graphiques des donnees quantitatives d'une dataframe pandas
    box: boite a moustaches
    pair: nuage de points bivarie
    cor: matrice de correlation
    cluster: double dendogramme
    histo: histogrammes
    
    """
    nrows=len(data.columns)%4
    quant_data=data[columns]
    counter=0
    # Creation de la matrice de nuages de points bivaries
    if mode is "pair":
        g=sns.pairplot(data[columns])
        g.fig.suptitle("Matrice de nuages de points bivariés")
    
    # Creation de la matrice de correlation        
    elif mode is "cor":
        corr = data[columns].corr()
        mask = np.triu(np.ones_like(corr, dtype=bool))
        # Set up the matplotlib figure
        f, ax = plt.subplots(figsize=(11, 9))
        # Generate a custom diverging colormap
        cmap = sns.diverging_palette(230, 20, as_cmap=True)
        # Draw the heatmap with the mask and correct aspect ratio
        g=sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
        g.fig.suptitle("Matrice de correlation")

    # Creation du double dendogramme
    elif mode is "cluster":
        corrmat = data[columns].corr(method='spearman')
        cg = sns.clustermap(corrmat, cmap="YlGnBu", linewidths=0.1);
    
    # Creation de graphiques univaries
    else:
        fig, ax = plt.subplots(4, nrows,figsize=(20,20))
        fig.suptitle('Quantitative variable univariate analysis')
        for r in range(nrows):
            for c in range(4):
                # Creation de boites a moustaches
                if mode is "box":
                    ax[c,r].boxplot(quant_data[columns[counter]])
                    ax[c,r].set_title(columns[counter])
                
                # Creation d'histogrammes
                elif mode is "histo":
                    ax[c,r].hist(quant_data[columns[counter]])
                    ax[c,r].set_title(columns[counter])

                counter+=1
                if counter==len(columns):
                    break
        
    return quant_data
            
quant_columns=list(data[data.columns[15:27]].columns) + ["additives_n"]
quant_columns
```

```python
quant_data=quant_data_exploration(data,quant_columns,mode="cor")
```

La plus grande correlation negative se trouve entre proteins_100g et sugars_100g

On voit d'apres le graph que les differents groupes d'aliments ont clairement des distributions differentes de nutriscores. On retrouve logiquement les fruits et legumes, ainsi que les cereales avec de moyennes de nutriscore inferieures a 0, consistant avec des aliments a haute qualite nutritionnelle.

```python
temp = data[["nutriscore_score","pnns_groups_1"]].groupby('pnns_groups_1')
temp2=[temp.get_group(x).reset_index(drop=True).iloc[:,0] for x in temp.groups]
columns=list(temp.groups.keys())
temp2=pd.concat(temp2,  ignore_index=True, axis=1)
temp2.columns=columns
temp2 = [temp2[col].dropna() for col in temp2]
f_val, p_val = stats.f_oneway(*temp2)
p_val
```

```python
res = stat()
res.anova_stat(df=data, res_var='nutriscore_score', anova_model='nutriscore_score ~ C(pnns_groups_1)')
res.anova_summary
```

Les analyses unidirectionnelles ANOVA indiquent que nous pouvons rejeter l'hypothese null que les moyennes de nutriscore sont egales a travers les categories d'aliments.

```python
res = stat()
res.tukey_hsd(df=data, res_var='nutriscore_score', xfac_var='pnns_groups_1', anova_model='nutriscore_score ~ C(pnns_groups_1)')
res.tukey_summary
```

```python
# res.anova_std_residuals are standardized residuals obtained from ANOVA (check above)
sm.qqplot(res.anova_std_residuals, line='45')
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Standardized Residuals")
plt.show()

# histogram
plt.hist(res.anova_model_out.resid, bins='auto', histtype='bar', ec='k') 
plt.xlabel("Residuals")
plt.ylabel('Frequency')
plt.show()
```
